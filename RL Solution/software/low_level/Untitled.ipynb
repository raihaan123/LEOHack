```python
import serial
ser = serial.Serial('COM3', 115200)
print(ser.name)
```

```python
def writeTwist(x, y, theta):
    sendString = "%f %f %f\n" % (x, y, theta)
    print(sendString)
    ser.write(bytes(sendString, "latin"))

writeTwist(0, 0, 0)
```

```python
import pygame
pygame.init()

def deadzone(value, deadzone):
    if(abs(value) < deadzone):
        return 0
    return value
 
def main():
    pygame.display.set_caption('JoyStick Example')
    surface = pygame.display.set_mode((800, 600))
    clock = pygame.time.Clock()
    running = True
 
    font = pygame.font.Font(None, 20)
    linesize = font.get_linesize()
    joysticks = [pygame.joystick.Joystick(i) for i in range(pygame.joystick.get_count())]
    for joy in joysticks:
        joy.init()
 
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.JOYAXISMOTION:
                #print("reading axis")
                axis0 = deadzone(joysticks[0].get_axis(0), 0.1)
                axis1 = deadzone(joysticks[0].get_axis(1), 0.1)
                axis2 = deadzone(joysticks[0].get_axis(2), 0.1)
                axis3 = deadzone(joysticks[0].get_axis(3), 0.1)
                
                #print("writing twist")
                writeTwist(-axis0 * 40.0, axis1 * 40.0, -axis3 / 5.0)
 
        surface.fill((0,0,0))
 
 
        pygame.display.flip()
        clock.tick(20)
 
    pygame.quit()
    writeTwist(0, 0, 0)
 
main()
```
